```python
import pandas as pd
import json
import requests
from pprint import pprint
import csv
```

```python
target_url = "https://data.cityofchicago.org/resource/6zsd-86xi.json"

crime_data = requests.get(target_url).json()

# Print the json
pprint(crime_data)


len(crime_data)

# Extract latitude and longitude
Block = crime_data["block"]
#lng = geo_data["results"][0]["geometry"]["location"]["lng"]
Block
```

```python
# reading csv files from 2013 to 2017 into a pandas data frame
crime_data_2013 = pd.read_csv("Resources/Crimes_-_2013.csv")
crime_data_2014 = pd.read_csv("Resources/Crimes_-_2014.csv")
crime_data_2015 = pd.read_csv("Resources/Crimes_-_2015.csv")
crime_data_2016 = pd.read_csv("Resources/Crimes_-_2016.csv")
crime_data_2017 = pd.read_csv("Resources/Crimes_-_2017.csv")
```

```python
crime_data = [crime_data_2013,crime_data_2014,crime_data_2015,crime_data_2016,crime_data_2017]
crime_data = pd.concat(crime_data,axis = 0)
crime_data
```

```python
# narrowing the table to relevant fields to use to be the geographical reference
crime_geo_data = crime_data[["Date", "Block", "Primary Type", "Community Area", "Latitude", "Longitude", "Year"]]
crime_geo_data.head()
```

```python
crime_geo_data.groupby("Year").count()
```

```python
crime_type = crime_geo_data.groupby(["Primary Type"]).count()
crime_type
```

```python
import matplotlib.pyplot as plt

#make graph of crimes per year
crime_data.Date = pd.to_datetime(crime_data.Date, format = '%m/%d/%Y %I:%M:%S %p')
crime_data.index = pd.DatetimeIndex(crime_data.Date)
plt.figure(figsize = (8,5))
crime_data.groupby([crime_data.index.year]).size().plot.bar()
plt.title('Crime Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Crimes Acts')
plt.show()
```

```python
#make graph of crimes per month
plt.figure(figsize = (8,5))
crime_data.groupby([crime_data.index.month]).size().plot.bar()
plt.title('Crime Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Crimes Acts')
plt.show()
```

```python
#graph of crimes per day
plt.figure(figsize = (8,5))
crime_data.groupby([crime_data.index.day]).size().plot.bar()
plt.title('Crime Per Day Of Month')
plt.xlabel('Day')
plt.ylabel('Number of Crimes Acts')
plt.show()
```

```python
#graph of crimes per hour
plt.figure(figsize = (8,5))
crime_data.groupby([crime_data.index.hour]).size().plot.bar()
plt.title('Crime Per Hour')
plt.xlabel('Hour')
plt.ylabel('Number of Crimes Acts')
plt.show()
```

```python
import numpy as np
crime_data_date = crime_data.pivot_table('ID', aggfunc = np.size, columns = 'Primary Type', 
                                         index = crime_data.index.date, fill_value = 0)
crime_data_date.index = pd.DatetimeIndex(crime_data_date.index)
#make graph of different types of crimes per month
Plot = crime_data_date.plot(figsize = (40,60), subplots = True, layout = (6,6),
                                  sharex = False, sharey = False)
plt.show()
```
